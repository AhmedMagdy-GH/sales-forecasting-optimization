```python
import pandas as pd
```

```python
file_path = '../../Data/Raw/'
date_column = ['Date']


hist_info_data_types = {
    'Store': 'int16',
    'DayOfWeek': 'int8',
    'Customers': 'int16',
    'Open': 'bool',
    'Promo': 'bool',
    'SchoolHoliday': 'bool',
    'StateHoliday': 'category',
    'Sales': 'int32'
}
store_data_types = {
    'Store': 'int16',
    'StoreType': 'category',
    'Assortment': 'category',
    'CompetitionDistance': 'float32',
    'CompetitionOpenSinceMonth': 'float32',
    'CompetitionOpenSinceYear': 'float32',
    'Promo2': 'bool',
    'Promo2SinceWeek': 'float32',
    'Promo2SinceYear': 'float32',
    'PromoInterval': 'string'
}
```

```python
df1 = pd.read_csv(
    file_path + 'historical_information.csv',
    parse_dates=date_column,
    dtype=hist_info_data_types,
    encoding="utf-8",
)
```

```python
df2 = pd.read_csv(
    file_path + 'store_details.csv',
    dtype=store_data_types,
    encoding="utf-8",
)

training_data = pd.merge(df1,df2, on='Store', how='left')
training_data.set_index("Date",inplace=True)

training_data.drop(columns="Customers", inplace=True)

training_data.head()
```

```python
training_data.to_csv("../../Data/Preprocessed/merged_training_data.csv")
```
